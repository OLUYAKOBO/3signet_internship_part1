#### Correlation Plot

```python
def corr():
    # Calculate the correlation matrix
    corr_matrix = df.corr()

    # Plot the heatmap
    plt.figure(figsize=(40, 40))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=2)

    plt.title('Correlation Matrix Heatmap')
    plt.show()
corr()
```

#### Chi Square test

```python
cat_cols
```

```python
contingency_table = pd.crosstab(df['Gender'], df['Marital status'])
print(contingency_table)
```

```python
import scipy.stats as stats

# Perform the Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

##### Principal component analysis
* In this section I am going to perform pca on the data to extract 10 features that best explain the dataset.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = scaler.fit_transform(x)

pca = PCA(n_components=10)  # Number of components you want to keep
principal_components = pca.fit_transform(df_standardized)

# Step 4: Create a DataFrame with principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2','PC3','PC4','PC5',
                                                         'PC6','PC7','PC8','PC9','PC10'])

# Step 5: Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
sum(pca.explained_variance_ratio_)
```

The sum of explained variance means that 65% of the info of the data is explained by the PCA
